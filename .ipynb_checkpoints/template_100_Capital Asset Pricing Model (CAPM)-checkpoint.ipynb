# CAPM - Capital Asset Pricing Model 

## simple linear regression으로 CAPM을 Modeling
- market portfolio 는 S&P500 사용
- Apple 단일 주식 portfolio 사용

### CAPM 공식 이용
### $ r_i = r_f + \beta * (r_m - r_f )$

```python
import pandas as pd
import seaborn as sns
import plotly.express as px
from copy import copy
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
stock_df = pd.read_csv("data/stock.csv")
stock_df.head()
```

날짜를 기준으로 데이터를 정렬하고 날짜를 인덱스로 설정

```python
stock_df = stock_df.sort_values(by='Date').set_index('Date')
stock_df.head()
```

```python
def normalize(df):
    x = df.copy()
    for col in x.columns:
        x[col] = x[col] / x[col][0]
    return x
```

```python
def interactive_plot(df, title):
    fig = px.line(title = title)
    for c in df.columns:
        fig.add_scatter(x = df.index, y = df[c], name = c)
    fig.show()
```

```python
interactive_plot(stock_df, "Prices")
```

price를 시작일 price로 normalize하는 함수

```python
interactive_plot(normalize(stock_df), 'Normalized 가격')
```

- 일일 수익률 계산

```python
daily_return = stock_df.pct_change()
daily_return.dropna(inplace=True)
daily_return
```

- 일일 평균 수익률

```python
daily_return.mean().sort_values(ascending=False)
```

### 단일 주식에 대한 베타 계산
- 단일 주식의 일일 수익률

```python
stock = "AAPL"
stock_name = "Apple"

daily_return[stock]
```

S&P500 (Market)의 일일 수익률

```python
daily_return['sp500']
```

선택한 주식과 S&P500(시장) 사이의 산점도를 플로팅합니다.

```python
daily_return.plot(kind='scatter', x='sp500', y=stock)
```

### Linear Regression을 이용한 BETA 계산

```python
X = daily_return['sp500'].values.reshape(-1, 1)   # 2차원 배열로 모양 변형 / -1 : 알아서 갯수를 계산하게끔
y = daily_return['AAPL']

X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.coef_, lr.intercept_   # Beta, 절편
```

알파는 시장을 이길 수 있는 전략의 능력을 나타냅니다(S&P500).  

알파는 "초과 수익률" 또는 "비정상적 수익률"을 나타내며 예를 들어 0.00175 알파는 포트폴리오의 수익률이 벤치마크 S&P500 지수를 0.17% 초과했음을 의미합니다.

```python
beta = lr.coef_[0]
alpha = lr.intercept_
beta, alpha
```

증권시장선 Line plot

```python
daily_return.plot(kind='scatter', x='sp500', y=stock)
plt.plot(daily_return['sp500'], beta * daily_return['sp500'] + alpha, 'r')
```

## beta를 구한 개별 주식에 CAPM 공식 적용
- 개별 주식의 기대 수익률 계산
### $ r_i = r_f + \beta * (r_m - r_f )$   

S&P500의 일일 평균 수익률

```python
daily_return['sp500'].mean()
```

S&P500의 연간 수익률

```python
rm = daily_return['sp500'].mean() * 252
```

개별 주식의 연간 수익률

```python
rf = 0
ri = rf + beta * (rm - rf)   # 현재 Apple 기준임
ri
```

```python
print(
f"""
CAPM 공식에 따르면,
단순히 마켓 인덱스에 투자할 때 연 {rm * 100:.2f}%의 수익이 기대되나, 
{stock_name} 주식에 투자할 때는 연 {ri * 100:.2f}%의 수익이 기대된다. 
공격적 투자.
""")
```

- AT&T의 기대 수익률 계산

```python
stock = 'T'
stock_name = 'AT&T'

X = daily_return['sp500'].values.reshape(-1, 1)
y = daily_return[stock]
X.shape, y.shape

lr = LinearRegression()
lr.fit(X, y)

beta = lr.coef_[0]

ri = rf + beta * (rm - rf)
ri
```

```python
print(
f"""
CAPM 공식에 따르면,
단순히 마켓 인덱스에 투자할 때 연 {rm * 100:.2f}%의 수익이 기대되나, 
{stock_name} 주식에 투자할 때는 연 {ri * 100:.2f}%의 수익이 기대된다. 
방어적 투자.
""")
```

## 모든 주식에 대한 베타 계산

```python
betas = {}
alphas = {}
X = daily_return['sp500'].values.reshape(-1, 1)

for col in daily_return.columns[:-1]:   # sp500 제외
    y = daily_return[col]
    lr = LinearRegression()
    lr.fit(X, y)
    beta = lr.coef_[0]
    alpha = lr.intercept_
    betas[col] = beta
    alphas[col] = alpha
    
    daily_return.plot(kind='scatter', x='sp500', y=col)
    plt.plot(daily_return['sp500'], beta * daily_return['sp500'] + alpha, 'r')
```

## CAMP 공식을 적용하여 포트폴리오 수익률 계산

모든 주식 목록을 얻습니다.

```python
stocks = list(betas.keys())
stocks
```

```python
ERs = {}
rf = 0
rm = daily_return['sp500'].mean() * 252

for stock in stocks:
    ERs[stock] = rf + betas[stock] * (rm - rf)
```

```python
ERs   # 연간 수익률
```

```python
print("CAPM 모델을 머신러닝으로 유추한 연간 기대 수익률")

for stock in stocks:
    print(f"{stock:<10} : {ERs[stock] * 100:.2f}%")
```

## Portfolio의 기대 수익 계산

- 위 주식들로 구성된 포트폴리오에서 동일한 가중치를 가정합니다.

```python
np.ones(8)   # 1로 되어있는 크기가 8인 numpy array 생성
```

```python
portfolio_weights = 1/len(stocks) * np.ones(len(stocks))
portfolio_weights
```

- 위 포트폴리오 기대수익률 계산

```python
sum(list(ERs.values()) * portfolio_weights) * 100   # 동일 비중 포트폴리오의 기대 수익률
```

- Apple 50%, MGM 50% 구성된 포트폴리오의 기대 수익률

```python
portfolio = ['AAPL', 'MGM']
portfolio_weights = 1/len(portfolio) * np.ones(len(portfolio))
portfolio_weights
```

```python
sum((0.5 * ERs['AAPL'] + 0.5 * ERs['MGM']) * portfolio_weights) * 100
```
