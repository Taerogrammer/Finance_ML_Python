# LSTM을 이용한 주가 예측

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

import yfinance as yf

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Lambda
```

- Apple 주식 가격 예측

```python
df = yf.download('AAPL', start='2010-01-01', end='2021-12-31', progress=False)
df.head()
```

```python
df.Close.plot()
```

LSTM layer로 하려고 하는 것은 지난 window-size 일(즉, t-window에서 t-1까지)의 역사적 가격을 사용하여 시간 t의 가격을 예측하는 것입니다. 정확한 가격이 아닌 추세를 파악하기 위해 노력할 것임을 명심하십시오.

`len(df) - 100`일의 데이터로 모델을 학습시키고 다음 100일의 데이터로 테스트하는 방식으로 데이터를 분할합니다.
### train/test split

```python
data_training = pd.DataFrame(df['Close'])[0:int(len(df)*0.7)]   # 학습용
data_test = pd.DataFrame(df['Close'])[int(len(df)*0.7):]   # test용
data_training.shape, data_test.shape
```

```python
scaler = MinMaxScaler()
data_training_array = scaler.fit_transform(data_training)
data_training_array.shape
```

```python
x_train = []
y_train = []
window = 100

for i in range(window, data_training.shape[0]):
    x_train.append(data_training_array[i-window: i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train.shape, y_train.shape
```

- data normalization

** train data가 잘 만들어 졌는지 확인하는 Tip** 

- hist의 각 element는 window개 timestep의 list입니다. 1씩 증가하기 때문에 X_train의 **두 번째 element의 마지막 항목**은 target의 **첫 번째 element**와 같아야 합니다. 또한 **마지막 element의 마지막 항목**은 target의 마지막 2번째 element와 같아야 합니다.

훈련 데이터의 i번째 원소의 마지막 값은 타깃 데이터의 (i-1)번째 값과 일치해야 합니다. 즉, X_train[i][-1] == y_train[i-1]이 성립해야 합니다. 이는 LSTM 모델이 'window' 개의 데이터를 바탕으로 그 다음 시점의 값을 예측하는 것이 목적이기 때문입니다.

또한, 훈련 데이터의 마지막 원소의 마지막 값은 타깃 데이터의 뒤에서 두 번째 값과 일치해야 합니다. 즉, X_train[-1][-1] == y_train[-2]가 성립해야 합니다. 이는 훈련 데이터와 타깃 데이터의 시퀀스가 올바르게 끝까지 유지되었음을 확인하는 검증입니다.

```python
x_train[1][-1][0] == y_train[0]   # True가 나오면 잘 만들어짐
```

```python
x_train[-1][-1][0] == y_train[-2]   # True가 나오면 잘 만들어짐
```

### model 생성 및 fitting

```python
model = Sequential()
model.add(LSTM(30, return_sequences=True, input_shape=(window, 1)))

# 과적합을 방지하기 위해 Dropout 사용
model.add(Dropout(0.2))

model.add(LSTM(60, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(120))
model.add(Dropout(0.2))

model.add(Dense(60))
model.add(Dense(1))   # Dense(1) : Output Layer
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(x_train, y_train, epochs=30, batch_size=32)
```

```python
plt.plot(history.history['loss'])
```

앞으로 100일 동안의 주가 추이를 예측해 보도록 하겠습니다. 우리의 목표는 실제 가격이 아니라 추세를 예측하는 것임을 기억하십시오. 첫번째 예측을 위해 과거 window 일 수 만큼의 data 가 필요하므로 training data의 마지막 window 갯수 만큼을 concatenate 합니다.

```python
past_window_days = data_training.tail(window)
final_df = pd.concat([past_window_days, data_test])   # test data 확인
final_df.shape
```

```python
input_data = scaler.fit_transform(final_df)
input_data
```

```python
x_test = []
y_test = []
window = 100

for i in range(window, final_df.shape[0]):
    x_test.append(input_data[i-window: i])
    y_test.append(input_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)
x_test.shape, y_test.shape
```

```python
x_test[1][-1][0] == y_test[0]   # True가 나오면 잘 만들어짐
```

```python
x_test[-1][-1][0] == y_test[-2]   # True가 나오면 잘 만들어짐
```

```python
y_predicted = model.predict(x_test)
y_predicted.shape
```

최근 일자 부분만 확대하여 시각화

```python
scaler.scale_, scaler.data_min_
```

```python
y_predicted = y_predicted / scaler.scale_ + scaler.data_min_   # 원래값으로 복원
```

```python
y_test = y_test / scaler.scale_ + scaler.data_min_
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, 'b', label='Original Data')
plt.plot(y_predicted, 'r', label = 'Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test[-20:], 'bx-', label='Original Data')
plt.plot(y_predicted[-20:], 'ro-', label = 'Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
```
